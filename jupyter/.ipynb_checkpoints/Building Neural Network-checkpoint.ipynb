```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, MaxPooling2D, Conv2D, Flatten, Conv3D, MaxPooling3D
from sklearn.preprocessing import LabelEncoder
import pickle
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, confusion_matrix
import cv2
import matplotlib.pyplot as plt
from tensorflow.keras.callbacks import TensorBoard
import seaborn as sns
```

### Load pickled data

```python
images = pickle.load(open('../pickles/images.pickle','rb'))
X = pickle.load(open('../pickles/X.pickle','rb'))
y = pickle.load(open('../pickles/y.pickle','rb'))
model = pickle.load(open('../pickles/model.pickle','rb'))
```

### Transoform y --> encoded array

```python
le = LabelEncoder()
y = np.array(y)
le.fit(y)
y_transform = le.transform(y)
```

### Scale X   [0,1]

```python
X = X/255
```

### Build nerual network

```python
model = Sequential()

model.add(Conv2D(64, (3,3), input_shape = X.shape[1:]))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64, (3,3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64, (3,3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(64))
model.add(Activation("relu"))

model.add(Dense(1))
model.add(Activation("sigmoid"))

model.compile(optimizer="adam",
              loss="binary_crossentropy",
              metrics=['accuracy'])

history = model.fit(X, y_transform, epochs=3, validation_split=0.1)
              
```

```python
model.summary()
```

```python
fig, axes = plt.subplots(1, 2, figsize = (12, 4))

sns.lineplot(x = range(len(history.history["loss"])), y = history.history["loss"], ax = axes[0], label = "Training Loss")
sns.lineplot(x = range(len(history.history["loss"])), y = history.history["val_loss"], ax = axes[0], label = "Validation Loss")

sns.lineplot(x = range(len(history.history["accuracy"])), y = history.history["accuracy"], ax = axes[1], label = "Training Accuracy")
sns.lineplot(x = range(len(history.history["accuracy"])), y = history.history["val_accuracy"], ax = axes[1], label = "Validation Accuracy")
axes[0].set_title("Loss"); axes[1].set_title("Accuracy")

sns.despine()
plt.show()
```

```python
pickle.dump(model, open('model.pickle','wb'))
```

### Inspect predictions

```python
X[]
```

```python
predictions = model.predict(X)
```

```python
results_df = pd.DataFrame({"image":images, "label":y, "pred_conf":predictions.reshape(1,-1)[0]})
```

```python
results_df['prediction'] = results_df['pred_conf'].apply(lambda x: "cat" if x < 0.5 else "dog")
```

```python
accuracy_score(results_df['label'],results_df['prediction'])
```

```python
confusion_matrix(results_df['label'],results_df['prediction'])
```

### View images misclassified

```python
wrong = results_df[results_df['label']!=results_df['prediction']].reset_index(drop=True)
```

```python
fig = plt.figure(1, figsize = (15, 15))

for i in range(25):
    plt.subplot(5, 5, i + 1)
    image = wrong.loc[i]['image']
    label = wrong.loc[i]['label']
    prediction = wrong.loc[i]['prediction']

    filepath = filepath = "data/unzipped/train/" + image
    image_array = cv2.imread(filepath)
    image_array = cv2.cvtColor(image_array, cv2.COLOR_BGR2RGB)
    plt.imshow(cv2.resize(image_array, (100,100)))
    plt.axis("off")
    plt.title(f"True Value: {label} \nPrediction: {prediction}")    
    
plt.tight_layout()
plt.show()
```
